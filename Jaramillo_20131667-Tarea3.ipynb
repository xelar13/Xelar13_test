4. El m√≥dulo vivienda incluye todos los hogares que entraron en el marco muestral, incluso aquellos que no participaron en la ENAHO por diversos motivos. Para eliminar estas viviendas, qued√©monos con aquellas cuyo ``` RESULT``` sea igual a 1 √≥ 2. Almacena este resultado para despu√©s.

````python
#Filtramos los datos de Vivienda para los que tienen Resulta 1 o 2, y lo guardamos en la variables nueva_viv
cond_1 = sel_viv['result'] == 1
cond_2 = sel_viv['result'] == 2

nueva_viv = sel_viv[cond_2 | cond_1]

nueva_viv
````

5. Queremos calcular una serie de variables (√≥ columnas nuevas):  
En sumaria: 
- Queremos el n√∫mero de miembros dependientes del hogar . Estos son el n√∫mero de personas que no perciben un ingreso √≥ la diferencia entre el total de miembros y el total de perceptores. 
- Queremos el ingreso mensual promedio por perceptor del hogar.
- Queremos la diferencia  entre ingreso y gasto del hogar

````python
#Calculamos la variable del n√∫mero de dependientes y la agregamos como nueva columna
sel_sum ['num_depn'] = (sel_sum['mieperho'] - sel_sum['percepho'])
sel_sum
````

````python
#Calculamos la variable del ingreso mensual promedio por perceptor del hogar y lo agregamos como una nueva columna
#Dividimos el ingreso anual entre 12 y entre el n√∫mero de perceptores
sel_sum ['ing_prom_per'] = (sel_sum['inghog2d'] / (sel_sum['percepho']*12)) 
sel_sum
````

````python
#La √∫ltima columna que nos piden es la diferencia entre el ingreso y gasto por hogar
sel_sum ['diferencia'] = (sel_sum['inghog2d'] - sel_sum['gashog2d'])
sel_sum
````

6. La variable de `pobreza`, en la sumaria, est√° codificada como integers 1, 2 y 3. Esto corresponde a:  
`1`: hogar pobre extremo  
`2`: hogar pobre no extremo  
`3`: hogar no  pobre  

En base a esta variable, crea dos variables m√°s:
1. Una variable que tenga `'pobre extremo'`, `'pobre no extremo'` y `'no pobre'` en vez  de 1, 2, 3.
2. Una variable que tenga `'pobre'` y `'no pobre'` (√≥ que englobe a los hogares pobres y pobres no extremos).



````python
# Creamos la variable de clasificaci√≥n de pobreza, para ello definimos una lista de condiciones y utilizamos numpy select
lista_cond1 = [
    (sel_sum['pobreza'] == 1) ,
    (sel_sum['pobreza'] == 2) ,
    (sel_sum['pobreza'] == 3)]
lista_clasif1 = ['Pobre Extremo', 'Pobre no extremo', 'No pobre']
sel_sum['claf_pobreza'] = np.select(lista_cond1, lista_clasif1, default='Not Specified')

sel_sum
````

````python
# Nos piden otra variable que englobe a Pobres Extremos y Pobres no Extremos en una variable Pobre, y al resto en No Pobre
lista_cond2 = [
    (sel_sum['pobreza'] == 1) ,
    (sel_sum['pobreza'] == 2) ,
    (sel_sum['pobreza'] == 3)]
lista_clasif2 = ['Pobre', 'Pobre', 'No pobre']
sel_sum['claf_pobreza2'] = np.select(lista_cond2, lista_clasif2, default='Not Specified')

sel_sum
````

7. Queremos saber el promedio de ingresos de los hogares (inghog2d) por nivel de pobreza. Haz un `groupby` para averiguar esto. 


(Por ejemplo, todos los hogares pobres extremos tendr√°n un √∫nico promedio, y as√≠ con las 2 categor√≠as de pobreza adicionales). 

````python
# Hacemos un groupby por ingresos y clasificaci√≥n de pobreza, adem√°s, lo ordenamos de menor a mayor
sel_sum[['inghog2d', 'claf_pobreza']].groupby('claf_pobreza').mean().sort_values('inghog2d').reset_index()
````

8. ¬°Ahora uniremos los datos!  
Pero antes un check de sanidad: Verifica que el n√∫mero  de filas de ambas bases de datos son las mismas (acu√©rdate que filtramos las observaciones por `result`).  La ENAHO recolecta informaci√≥n de estos 34,490 hogares a trav√©s de todos sus m√≥dulos. Para utilizar estos m√≥dulos a la vez, realizaremos una "uni√≥n" (merge).   

Une ambas bases con un _inner merge_. Recuerda que el id que identifica a cada hogar se compone de ` conglome vivienda hogar`.  

````python
#Hacemos un merge entre sumaria y viviendas -las cuales fueron filtradas anteriormente seg√∫n result-. Esta nueva base se llama BaseFinal
BaseFinal = pd.merge(sel_sum,
                     nueva_viv,
                    left_on = ('conglome', 'vivienda', 'hogar'),
                    right_on = ('conglome', 'vivienda', 'hogar'),
                    how = 'inner')
BaseFinal
````

9. Haz un ```groupby``` que agrupe (`agg`) los datos por nivel de pobreza  y la nbi1. Queremos hallar el promedio y la mediana de los ingresos (inghog2d) y los gastos (gashog2d) por el producto cartesiano de las categor√≠as de estas 2 variables - pobreza y nbi 1 -. 

El producto cartesiano es tan s√≥lo todas las posibles combinaciones entre categor√≠as. 
En este caso, tendremos:
- pobre extremo con nbi1
- pobre extremo sin nbi1
- pobre no extremo con nbi1
- pobre no extremo sin nbi1
- no pobre con nbi1
- no pobre sin nbi1

(Este es un  ejemplo de c√≥mo, al cruzar un indicador de pobreza monetaria -pobreza- y pobreza multidimensional-nbis-, podemos seguir observando brechas).

````python
#Definimos a la variable prodcartesiano seg√∫n lo que se nos solicita en la pregunta

prodcartesiano = BaseFinal.groupby (['claf_pobreza', 'nbi1']).agg(
    {'inghog2d': ['mean', 'median'], 'gashog2d': ['mean', 'median']}).reset_index().sort_values('claf_pobreza')


prodcartesiano
````

10. Crea una variable que nos informe sobre los hogares que tienen las 5 nbis. Esta tomar√° el valor de 1 si tiene las 5 nbis, 0 en caso contrario (as√≠  tenga 4, 3). 

Calcula el promedio de miembros dependientes por cada categor√≠a de esta nueva variable. 

````python
#Convertimos los datos string a integer de las NBI
BaseFinal['nbi1'] = pd.to_numeric(BaseFinal['nbi1'], downcast='integer')
BaseFinal['nbi2'] = pd.to_numeric(BaseFinal['nbi2'], downcast='integer')
BaseFinal['nbi3'] = pd.to_numeric(BaseFinal['nbi3'], downcast='integer')
BaseFinal['nbi4'] = pd.to_numeric(BaseFinal['nbi4'], downcast='integer')
BaseFinal['nbi5'] = pd.to_numeric(BaseFinal['nbi5'], downcast='integer')

#Primero creamos la nueva variable que dar√° 1 si el hogar posee las 5 NBI y 0 en otro caso. Para esto utilizamos apply lambda

BaseFinal['CincoNBI'] = BaseFinal.apply(lambda row: 1 
                                        if  (row.nbi1 == 1) & (row.nbi2 == 1) & (row.nbi3 == 1) & (row.nbi4 == 1) & (row.nbi5 == 1)
                                        else 0, axis = 1)
BaseFinal
````

````python
# Esta l√≠nea de c√≥digo es para comprobar cu√°ntos hogares tienen 5 NBI. Como se comprueba, ning√∫n hogar posee las cinco NBI
 
BaseFinal['CincoNBI'].value_counts()

# Los resultados son correctos, ya que se corrobor√≥ a trav√©s de una exportaci√≥n a Excel de la BaseFinal si hab√≠a un hogar
# Con las 5 NBI y el resultado fue el mismo que en Python
````

````python
# Ahora calculamos el promedio de miembros dependientes que nos solicitan

BaseFinal[['num_depn', 'CincoNBI']].groupby('CincoNBI').mean().reset_index()

# Como observamos -y por lo dicho anteriormente- el promedio de hogares con las cinco NBI es nulo, ya que ning√∫n hogar cumple
# con dicha condici√≥n.
# Solo tenemos un promedio de n√∫mero de dependoetnes para los hogares que no tienen las cinco NBI
````
