4. El módulo vivienda incluye todos los hogares que entraron en el marco muestral, incluso aquellos que no participaron en la ENAHO por diversos motivos. Para eliminar estas viviendas, quedémonos con aquellas cuyo ``` RESULT``` sea igual a 1 ó 2. Almacena este resultado para después.

5. Queremos calcular una serie de variables (ó columnas nuevas):  
En sumaria:

6. La variable de `pobreza`, en la sumaria, está codificada como integers 1, 2 y 3. Esto corresponde a:  
`1`: hogar pobre extremo  
`2`: hogar pobre no extremo  
`3`: hogar no  pobre  

En base a esta variable, crea dos variables más:
1. Una variable que tenga `'pobre extremo'`, `'pobre no extremo'` y `'no pobre'` en vez  de 1, 2, 3.
2. Una variable que tenga `'pobre'` y `'no pobre'` (ó que englobe a los hogares pobres y pobres no extremos).



````python
# Creamos la variable de clasificación de pobreza, para ello definimos una lista de condiciones y utilizamos numpy select
lista_cond1 = [
    (sel_sum['pobreza'] == 1) ,
    (sel_sum['pobreza'] == 2) ,
    (sel_sum['pobreza'] == 3)]
lista_clasif1 = ['Pobre Extremo', 'Pobre no extremo', 'No pobre']
sel_sum['claf_pobreza'] = np.select(lista_cond1, lista_clasif1, default='Not Specified')

sel_sum
````

````python
# Nos piden otra variable que englobe a Pobres Extremos y Pobres no Extremos en una variable Pobre, y al resto en No Pobre
lista_cond2 = [
    (sel_sum['pobreza'] == 1) ,
    (sel_sum['pobreza'] == 2) ,
    (sel_sum['pobreza'] == 3)]
lista_clasif2 = ['Pobre', 'Pobre', 'No pobre']
sel_sum['claf_pobreza2'] = np.select(lista_cond2, lista_clasif2, default='Not Specified')

sel_sum
````

7. Queremos saber el promedio de ingresos de los hogares (inghog2d) por nivel de pobreza. Haz un `groupby` para averiguar esto. 


(Por ejemplo, todos los hogares pobres extremos tendrán un único promedio, y así con las 2 categorías de pobreza adicionales). 

````python
# Hacemos un groupby por ingresos y clasificación de pobreza, además, lo ordenamos de menor a mayor
sel_sum[['inghog2d', 'claf_pobreza']].groupby('claf_pobreza').mean().sort_values('inghog2d').reset_index()
````

8. ¡Ahora uniremos los datos!  
Pero antes un check de sanidad: Verifica que el número  de filas de ambas bases de datos son las mismas (acuérdate que filtramos las observaciones por `result`).  La ENAHO recolecta información de estos 34,490 hogares a través de todos sus módulos. Para utilizar estos módulos a la vez, realizaremos una "unión" (merge).   

Une ambas bases con un _inner merge_. Recuerda que el id que identifica a cada hogar se compone de ` conglome vivienda hogar`.  

````python
#Hacemos un merge entre sumaria y viviendas -las cuales fueron filtradas anteriormente según result-. Esta nueva base se llama BaseFinal
BaseFinal = pd.merge(sel_sum,
                     nueva_viv,
                    left_on = ('conglome', 'vivienda', 'hogar'),
                    right_on = ('conglome', 'vivienda', 'hogar'),
                    how = 'inner')
BaseFinal
````

9. Haz un ```groupby``` que agrupe (`agg`) los datos por nivel de pobreza  y la nbi1. Queremos hallar el promedio y la mediana de los ingresos (inghog2d) y los gastos (gashog2d) por el producto cartesiano de las categorías de estas 2 variables - pobreza y nbi 1 -. 

El producto cartesiano es tan sólo todas las posibles combinaciones entre categorías. 
En este caso, tendremos:
- pobre extremo con nbi1
- pobre extremo sin nbi1
- pobre no extremo con nbi1
- pobre no extremo sin nbi1
- no pobre con nbi1
- no pobre sin nbi1

(Este es un  ejemplo de cómo, al cruzar un indicador de pobreza monetaria -pobreza- y pobreza multidimensional-nbis-, podemos seguir observando brechas).

````python
#Definimos a la variable prodcartesiano según lo que se nos solicita en la pregunta

prodcartesiano = BaseFinal.groupby (['claf_pobreza', 'nbi1']).agg(
    {'inghog2d': ['mean', 'median'], 'gashog2d': ['mean', 'median']}).reset_index().sort_values('claf_pobreza')


prodcartesiano
````

10. Crea una variable que nos informe sobre los hogares que tienen las 5 nbis. Esta tomará el valor de 1 si tiene las 5 nbis, 0 en caso contrario (así  tenga 4, 3). 

Calcula el promedio de miembros dependientes por cada categoría de esta nueva variable. 

````python
#Convertimos los datos string a integer de las NBI
BaseFinal['nbi1'] = pd.to_numeric(BaseFinal['nbi1'], downcast='integer')
BaseFinal['nbi2'] = pd.to_numeric(BaseFinal['nbi2'], downcast='integer')
BaseFinal['nbi3'] = pd.to_numeric(BaseFinal['nbi3'], downcast='integer')
BaseFinal['nbi4'] = pd.to_numeric(BaseFinal['nbi4'], downcast='integer')
BaseFinal['nbi5'] = pd.to_numeric(BaseFinal['nbi5'], downcast='integer')

#Primero creamos la nueva variable que dará 1 si el hogar posee las 5 NBI y 0 en otro caso. Para esto utilizamos apply lambda

BaseFinal['CincoNBI'] = BaseFinal.apply(lambda row: 1 
                                        if  (row.nbi1 == 1) & (row.nbi2 == 1) & (row.nbi3 == 1) & (row.nbi4 == 1) & (row.nbi5 == 1)
                                        else 0, axis = 1)
BaseFinal
````

````python
# Esta línea de código es para comprobar cuántos hogares tienen 5 NBI. Como se comprueba, ningún hogar posee las cinco NBI
 
BaseFinal['CincoNBI'].value_counts()

# Los resultados son correctos, ya que se corroboró a través de una exportación a Excel de la BaseFinal si había un hogar
# Con las 5 NBI y el resultado fue el mismo que en Python
````

````python
# Ahora calculamos el promedio de miembros dependientes que nos solicitan

BaseFinal[['num_depn', 'CincoNBI']].groupby('CincoNBI').mean().reset_index()

# Como observamos -y por lo dicho anteriormente- el promedio de hogares con las cinco NBI es nulo, ya que ningún hogar cumple
# con dicha condición.
# Solo tenemos un promedio de número de dependoetnes para los hogares que no tienen las cinco NBI
````
