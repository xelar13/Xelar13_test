   Altair también hace otros tipos de gráficos estadísticos. Por ejemplo un diagrama de caja o [_boxplot_](https://altair-viz.github.io/gallery/boxplot.html). Con el ejemplo del link y con la base de datos del IDH distrital de la T4, haz un diagrama de caja donde el eje x sea la variable dummy `Huancavelica`, el eje y, cualquier variable contínua (puedes modificar el width/ height para que el gráfico luzca mejor). 

```python
#Importamos los paquetes de pandas y Altair
import pandas as pd
import altair as alt
```

```python
# Leemos la base de datos del IDH distrital
IDH_distrital = pd.read_csv("idh_distrital_2019.csv",encoding = "latin-1")
IDH_distrital
```

```python
# Realizamos el gráfico solicitado tomando a Huancavelica como variable nominal N, y a la
# población con secundaria completa como variable continua Q
# Modificamos la longitud y ancho del gráfico para que se aprecie mejor
alt.Chart(IDH_distrital).mark_boxplot(color="#FF338F").encode(
    x='Huancavelica:N',
    y=alt.Y('pob_secundaria_completa:Q', title="Población con Secundaria Completa"
)).properties(width=350, height=400)
```
